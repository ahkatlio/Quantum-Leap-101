# Qubits: Building blocks of the Quantum Computing
## Single Qubit Representation
A single qubit is the fundamental unit of quantum information, analogous to a classical bit but with unique quantum properties. It can exist in a superposition of states, represented mathematically using Dirac notation.

A single qubit can be expressed as:
$$|\psi\rangle = \alpha|0\rangle + \beta|1\rangle$$
where:

- $|0\rangle$ and $|1\rangle$ are the basis states (**the computational basis**).
- $\alpha$ and $\beta$ are complex coefficients (**probability amplitudes**) that satisfy the normalization condition.

The basis states $|0\rangle$ and $|1\rangle$ can be represented as column vectors:
$$|0\rangle = \begin{bmatrix} 1\\ 0 \end{bmatrix},\text{ } |1\rangle = \begin{bmatrix} 0\\ 1 \end{bmatrix} $$
A general qubit state $|\psi\rangle$ can then be represented as a column vector:
$$ |\psi\rangle = \alpha \begin{bmatrix} 1\\ 0 \end{bmatrix}+ \beta \begin{bmatrix} 0\\ 1 \end{bmatrix}  = \begin{bmatrix} \alpha\\ \beta \end{bmatrix} $$
## Multi-Qubit Representation
Multi-qubit systems are essential for realizing the full potential of quantum computing, enabling exponential state representation, entanglement, parallelism, and complex operations that significantly enhance computational capabilities beyond what single-qubit systems can achieve. To create a multi-qubit system, we use the **tensor product** of the individual qubit states.

For example, if we have two qubits, $|\psi_1\rangle$ and $|\psi_2\rangle$:
$$\begin{aligned}
|\psi_1\rangle &= \alpha_1 |0\rangle + \beta_1 |1\rangle \\
|\psi_2\rangle &= \alpha_2 |0\rangle + \beta_2 |1\rangle
\end{aligned}
$$
The combined state of the two qubits is given by the tensor product:
$$\begin{aligned}
|\Psi\rangle = |\psi_1\rangle \otimes |\psi_2\rangle &= (\alpha_1 |0\rangle + \beta_1 |1\rangle) \otimes (\alpha_2 |0\rangle + \beta_2 |1\rangle)\\
&= \alpha_1\alpha_2|00\rangle+ \alpha_1\beta_2|01\rangle+\beta_1\alpha_2|10\rangle+ \beta_1\beta_2 |11\rangle
\end{aligned}
$$
### Generalization to n Qubits:
Each qubit can exist in one of two states: $|0\rangle$ or $|1\rangle$. Therefore, for $n$ qubits, the total number of distinct combinations of these states is $2^n$. This is because each qubit doubles the number of possible states:

- 1 qubit: $2^1$ = 2 states ($|0\rangle$,$|1\rangle$)
- 2 qubits: $2^2$ = 4 states ($|00\rangle$,$|01\rangle$,$|10\rangle$,$|11\rangle$)
- 3 qubits: $2^3$ = 8 states ($|000\rangle$,$|001\rangle$,$|010\rangle$,$|011\rangle$,$|100\rangle$,$|101\rangle$,$|110\rangle$,$|111\rangle$)
- And so on...

The state for a system of $n$ qubits can be represented as:
$$|\Psi\rangle = \sum_{i=0}^{2^{n}-1}c_i|i\rangle$$
Where:

- $|i\rangle$ represents each possible combination of the basis states. That means each integer corresponds to a unique binary representation that indicates the state of each qubit. For example,
  - For $n=3$ (3 qubits), the states can be indexed as follows:
    - 0 corresponds to $|000\rangle$
    - 1 corresponds to $|001\rangle$
    - 2 corresponds to $|010\rangle$
    - 3 corresponds to $|011\rangle$
    - 4 corresponds to $|100\rangle$
    - 5 corresponds to $|110\rangle$
    - 6 corresponds to $|111\rangle$
- The coefficients $c_i$ represent the **probability amplitudes** associated with each of the $2^n$ states.

The normalization condition $\sum |c_i|^2=1$ ensures that the total probability of measuring the qubit system in any of its possible states sums to 1.
## Completeness Relation
Let's say we have orthonormal basis |0⟩ and |1⟩. The state of a qubit in this basis can be expressed as:
$$|\psi\rangle = \alpha |a\rangle + \beta |b\rangle$$
Here α and β ∈ ℂ which are scalar.
As |0⟩ and |1⟩ are orthonormal basis:
$$\langle a | a \rangle = 1, \langle b | b \rangle = 1$$
$$\langle a | b \rangle = 0, \langle b | a \rangle = 0$$
Interestingly, α can also be expressed as $\langle a | \psi\rangle$

You might be wondering how this works. Let's break it down!
$$\begin{aligned}
\langle a | \psi\rangle &= \langle a | ( \alpha |a\rangle + \beta |b\rangle)\\
&= \alpha \langle a | a\rangle + \beta \langle a | b\rangle\\
&= \alpha \times 1 + \beta \times 0\\
&= \alpha
\end{aligned}
$$
Similarly β can also be expressed as $\langle b | \psi\rangle$
Let's break it down!
$$\begin{aligned}
\langle b | \psi\rangle &= \langle b | ( \alpha |a\rangle + \beta |b\rangle)\\
&= \alpha \langle b | a\rangle + \beta \langle b | b\rangle\\
&= \alpha \times 0 + \beta \times 1\\
&= \beta
\end{aligned}
$$
So the qubit can also be expressed as:
$$|\psi\rangle = \langle a | \psi\rangle |a\rangle + \langle b | \psi\rangle |b\rangle$$
As indicated, the inner products are just scalars. Therefore, instead of multiplying them by the vectors $|a\rangle$ and $|b\rangle$ on the left, we can equivalently multiply them on the right:
$$\begin{aligned}
|\psi\rangle &= |a\rangle\langle a||\psi\rangle + |b\rangle\langle b||\psi\rangle\\
&= ( |a\rangle\langle a|+|b\rangle\langle b|)|\psi\rangle\\
\end{aligned}
$$
For this to be true for all $|\psi\rangle$, we must have:
$$|a\rangle\langle a|+|b\rangle\langle b| = \mathbf{I}$$
This is known as the completeness relation, which indicates that the state of any qubit can be expressed in terms of the basis vectors $|a\rangle$ and $|b\rangle$. This property is referred to as completeness.
