# Measurement
In quantum computing, measurement is a fundamental operation that extracts information from a quantum system. When we actually measure a qubit, the outcome is either 0 or 1. If we measure a qubit in the state $α |0\rangle + \beta |1\rangle$, then the probability that the outcome is 0 is $|\alpha|^2$, and the probability of the outcome of 1 is $|\beta|^2$.
Moreover, the act of measuring fundamentally changes the state. If the outcome of the measurement is 0 then after the measurement the electron is actually in the ground state or |0⟩ and if  the outcome is 1 the new state is |1⟩. This is the **principle of measurement**, and it accounts for some of the strangeness of quantum mechanics.
## Measurement Operators
In quantum mechanics, every measurable quantity (observable) is associated with a **linear Hermitian operator**. For example, if we want to measure the spin of a qubit, the corresponding observable could be the spin in the $z$ direction, denoted as $S_z$.

The measurement operator $M_m$ is defined based on the observable being measured. If we denote the measurement outcomes as $m$ (which could be eigenvalues of the observable), the operator corresponding to each outcome can be constructed as follows:

- $M_0 = |0\rangle \langle 0|$ for measuring the outcome corresponding to spin-up (or state ∣0⟩).
- $M_1 = |1\rangle \langle 1|$ for measuring the outcome corresponding to spin-down (or state |1⟩)

#### Properties of Measurement Operators
- $M_m$ is Hermitian: $M_m = M_m^†$. This property ensures that the eigenvalues (measurement outcomes) are real numbers.
- $M_0$ and $M_1$ are projection operators: They project the state of the system onto the corresponding eigenstate when a measurement is made. $M_0^2= M_0$ and $M_1^2= M_1$

#### Calculating Probabilities
To find the probability of obtaining a specific measurement outcome $m$, we use the formula:
$$ \begin{aligned}
p(m) &= \langle \psi| M_m M_m^\dagger |\psi \rangle\\
&= \langle \psi| M_m |\psi \rangle
\end{aligned}
$$
here, $|\psi\rangle$ is the state of the qubit before measurement. It can be a superposition of states, such as $|\psi\rangle = \alpha |0\rangle + \beta |1\rangle$.

For our defined operators:
$$ \begin{aligned}
p(0) &= \langle \psi| M_0 M_0^\dagger |\psi \rangle\\
&= \langle \psi| M_0 M_0 |\psi \rangle\\
&= \langle \psi| |0\rangle\langle0|\cdot |0\rangle \langle 0||\psi\rangle \\
&= \langle \psi| |0\rangle\langle0 |0\rangle \langle 0||\psi\rangle \\
&= \langle \psi|0\rangle \langle 0|\psi\rangle\\
&= (\alpha^* \langle 0|0\rangle + \beta^* \langle 1|0\rangle)(\alpha \langle 0|0\rangle + \beta \langle 0|1\rangle) \\
&= \alpha^* \alpha\\
&= |\alpha|^2\\
\end{aligned}
$$

$$ \begin{aligned}
p(1) &= \langle \psi| M_1 M_1^\dagger |\psi \rangle\\
&= \langle \psi| M_1 M_1 |\psi \rangle\\
&= \langle \psi| |1\rangle\langle 1|\cdot |1\rangle \langle 1||\psi\rangle \\
&= \langle \psi| |1\rangle\langle 1 |1\rangle \langle 1||\psi\rangle \\
&= \langle \psi|1\rangle \langle 1|\psi\rangle \\
&= (\alpha^* \langle 0|1\rangle + \beta^* \langle 1|1\rangle)(\alpha \langle 1|0\rangle + \beta \langle 1|1\rangle) \\
&= \beta^* \beta\\
&= |\beta|^2\\
\end{aligned}
$$

Note:

We can write $\langle \psi|1\rangle \langle 1|\psi\rangle $ as: $|\langle 1|\psi\rangle|^2$. Also we can write $\langle \psi|0\rangle \langle 0|\psi\rangle $ as $|\langle 0|\psi\rangle|^2 $

## Sequantial Single-Qubit Measurements
What happens if we measure the qubits, one after another? Let's say we have two qubits in the state:
$$|\psi_{pre}\rangle = \frac{1}{\sqrt{2}}|00\rangle+\frac{1}{2}|01\rangle+\frac{\sqrt{3}}{4}|10\rangle+\frac{1}{4}|11\rangle$$
If we first measure the left qubit, we get
- $|0\rangle$ with probability $\frac{3}{4}$, and the state collapses to $|\psi_0\rangle = \sqrt{\frac{2}{3}}|00\rangle+\frac{1}{\sqrt{3}}|01\rangle$
- $|1\rangle$ with probability $\frac{1}{4}$, and the state collapses to $|\psi_1\rangle = \frac{\sqrt{3}}{2}|10\rangle + \frac{1}{2}|11\rangle$

Now if we measure the right qubit after this, the probability of getting $|00\rangle$ is the probability of first getting $|0\rangle$ for the left qubit, which was $\frac{3}{4}$, times the probability of getting $|0\rangle$ for the right qubit, which is $\frac{2}{3}$ because the state collapsed after the first measurement. We can perform this calculation for every possible outcome:
- $P(00) = \left(|\langle00|\psi_{pre}\rangle\right|^2+|\langle 01|\psi_{pre}\rangle|^2)(|\langle 00|\psi_0\rangle|^2) = \frac{3}{4} \times \frac{2}{3}= \frac{1}{2} $
- $P(01) = \left(|\langle00|\psi_{pre}\rangle\right|^2+|\langle 01|\psi_{pre}\rangle|^2)(|\langle 01|\psi_0\rangle|^2) = \frac{3}{4} \times \frac{1}{3}= \frac{1}{4} $
- $P(10) = \left(|\langle10|\psi_{pre}\rangle\right|^2+|\langle 11|\psi_{pre}\rangle|^2)(|\langle 10|\psi_1\rangle|^2) = \frac{1}{4} \times \frac{3}{4}= \frac{3}{16} $
- $P(11) = \left(|\langle10|\psi_{pre}\rangle\right|^2+|\langle 11|\psi_{pre}\rangle|^2)(|\langle 11|\psi_1\rangle|^2) = \frac{1}{4} \times \frac{1}{4}= \frac{1}{16} $

Note that: Measuring both qubits is the same as measuring one after another.

## Measurement-Based Quantum Computation
Measurement-based quantum computation (MBQC) is a model of quantum computing where the computation is driven by measurements on a highly entangled resource state, such as a cluster state. By performing a sequence of single-qubit measurements, any quantum circuit can be executed.
