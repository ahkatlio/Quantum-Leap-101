# Entanglement
Quantum entanglement is a bizarre phenomenon in quantum mechanics where two or more particles become linked, such that the state of one particle depends on the state of the others, even if they are separated by large distances. This connection persists regardless of how far apart the particles are, which Einstein famously described as "spooky action at a distance."

In quantum computing, when qubits become entangled, the state of one qubit becomes dependent on the state of another, regardless of the distance separating them. This interconnectedness allows quantum computers to perform complex calculations more efficiently than classical computers. Entanglement allows parallelism in computation, meaning that operations on entangled qubits can yield results that would require exponentially more resources in a classical system. For example, manipulating one qubit in an entangled pair can instantaneously affect its partner, enabling faster information processing.

## Example: Bell States
A well-known example of entanglement in quantum computing involves **Bell states**, which are a set of four specific quantum states of two qubits that are maximally entangled. The four Bell states are:

1. $|\Phi^+\rangle = \frac{1}{\sqrt{2}}(|00\rangle +|11\rangle)$
2. $|\Phi^-\rangle = \frac{1}{\sqrt{2}}(|00\rangle -|11\rangle)$
3. $|\Psi^+\rangle = \frac{1}{\sqrt{2}}(|01\rangle +|10\rangle)$
4. $|\Psi^-\rangle = \frac{1}{\sqrt{2}}(|01\rangle -|10\rangle)$

These states show how the measurement of one qubit affects the other. For example, if two qubits are in the state $|\Phi^+\rangle$ and one qubit is measured to be 0, the other qubit will also be found to be 0. Similarly, if the first qubit is measured to be 1, the second will also be 1. This correlation persists regardless of the physical distance between the qubits. Later in this series we will talk about the Bell states in details.

## Why Entangled States Cannot Be Represented as a Tensor Product of Two Qubits
Entanglement cannot be expressed as a simple tensor product of two qubit states.While the tensor product is a mathematical operation used to combine quantum states, it is essential to distinguish between **separable** (non-entangled) states and **entangled states**.

A state is considered **separable** (or non-entangled) if it can be expressed as a tensor product of its constituent qubit states. For example, a state like:

$$|00\rangle = |0\rangle\otimes|0\rangle$$
is separable.
That means, it represents situations where the qubits are independent of each other.

Conversely, an **entangled state** cannot be written as a tensor product of individual qubit states. For example, the Bell state:

$$|\Phi^+\rangle = \frac{1}{\sqrt{2}}(|00\rangle +|11\rangle)$$

Can you say why?

### Let's do the math:
Let's say, $|\Phi^\pm\rangle = (\alpha_1|0\rangle+\beta_1|1\rangle)\otimes(\alpha_2|0\rangle+\beta_2|1\rangle)$
$$\begin{aligned}&\therefore \frac{1}{\sqrt{2}}(|00\rangle\pm|11\rangle) = (\alpha_1|0\rangle+\beta_1|1\rangle)\otimes(\alpha_2|0\rangle+\beta_2|1\rangle)\\
&\implies \frac{1}{\sqrt{2}}\begin{bmatrix} 1 \\ 0 \\ 0 \\ 0 \end{bmatrix} \pm \begin{bmatrix}0\\0\\0\\1\end{bmatrix} = \begin{bmatrix}\alpha_1\\\beta_1\end{bmatrix}\otimes\begin{bmatrix}\alpha_2\\\beta_2\end{bmatrix}\\
&\implies \begin{bmatrix}\frac{1}{\sqrt{2}}\\0\\0\\\pm\frac{1}{\sqrt{2}}\end{bmatrix} = \begin{bmatrix}\alpha_1\alpha_2\\\alpha_1\beta_2\\\beta_1\alpha_2\\\beta_1\beta_2\end{bmatrix}\\  \end{aligned}$$
$$\begin{aligned}\therefore \alpha_1\alpha_2 &= \frac{1}{\sqrt{2}}\\ \alpha_1\beta_2 &= 0 \\ \beta_1\alpha_2 &= 0 \\ \beta_1\beta_2\ &= \pm \frac{1}{\sqrt{2}}\end{aligned}$$

Let's analyze these conditions:

- **Condition 1:** $\alpha_1\alpha_2 = \frac{1}{\sqrt{2}}$ indicates that the product of the amplitudes of the $|0\rangle$ states from both qubits must equal $\frac{1}{\sqrt{2}}$
- **Condition 2:** $\alpha_1\beta_2 = 0$ implies that either $\alpha_1 = 0$ or $\beta_2 = 0$. This means that one of the two states in the second qubit must not contribute to the superposition.
- **Condition 3:** $\beta_1\alpha_2 = 0$  implies that either $\beta_1 = 0$ or $\alpha_2=0$. This means that one of the two states in the first qubit must not contribute to the superposition.
- **Condition 4:** $ \beta_1\beta_2\ = \pm \frac{1}{\sqrt{2}}$ indicates that the product of the amplitudes of the $|1\rangle$ states from both qubits must equal $\pm\frac{1}{\sqrt{2}}$

Given the conditions, let's consider the implications:

- From **Condition 2** and **Condition 3**, at least one of $\alpha_1$ or $\beta_2$ must be zero, and at least one of $\beta_1$ or $\alpha_2$  must be zero. This means that at least one amplitude from each qubit's state must be zero.
- If, for example, $\alpha_1 = 0$  (from Condition 2), then $\alpha_2$ must be such that $\alpha_1\alpha_2 = \frac{1}{\sqrt{2}}$ cannot hold, since $0 \times \alpha_2 = 0 \neq \frac{1}{\sqrt{2}}$.
- Similarly, if $\beta_1 = 0$  (from Condition 3), then $\beta_2$ must also be such that $\beta_1\beta_2\ = \pm \frac{1}{\sqrt{2}}$ cannot hold, since $0 \times \beta_2 = 0 \neq \pm\frac{1}{\sqrt{2}}$

Thus, it is impossible for both $\alpha_1\alpha_2 = \frac{1}{\sqrt{2}}$ and $\beta_1\beta_2\ = \pm \frac{1}{\sqrt{2}}$ to hold true simultaneously while also satisfying the conditions that force at least one of the coefficients in each qubit's state to be zero. Therefore, these conditions cannot represent a valid quantum state in the context of separability and entanglement.

## Monogamy of Entanglement

The monogamy of entanglement is a rule in quantum mechanics that says: if two particles are strongly entangled, they can't be equally entangled with a third particle. It's like an "exclusive relationship" in the quantum world.

### Classical Correlations (Normal Relationships):
 - In classical physics, particles can share information freely with many others. For example, if particle A is correlated with B, it can still be correlated with C without any problem.
 - Think of it like friends sharing gossip—everyone can know the same thing!

### Quantum Entanglement (Exclusive Relationships):
Say there are three people: Alice, Bob, and Charlie. They have their own qubits. Is it possible for them to correlate among each other? If Alice and Bob are maximally entangled, they cannot be entangled with Charlie at the same time. This means that the entanglement shared between Alice and Bob excludes Charlie from being part of that strong correlation. This principle ensures that in a three-party quantum system, the total amount of entanglement is limited. For instance, if Alice and Bob share a strong entangled state, any attempt to create a similar entangled relationship with Charlie will reduce the entanglement between Alice and Bob
